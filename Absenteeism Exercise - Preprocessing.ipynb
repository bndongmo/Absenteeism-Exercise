```python
# Imports
import pandas as pd
import numpy as np
```

```python
# Load data
absenteeism_data = pd.read_csv('absenteeism_data.csv')
absenteeism_data.head()
```

```python
# Drop ID column
df_preprocessed = absenteeism_data.drop('ID', axis=1)
df_preprocessed.head()
```

```python
# Group reasons together

df_preprocessed.insert(0, 'Reason_4', np.where((df_preprocessed['Reason for Absence'] >= 22) & (df_preprocessed['Reason for Absence'] <= 28),1,0))
df_preprocessed.insert(0, 'Reason_3', np.where((df_preprocessed['Reason for Absence'] >= 18) & (df_preprocessed['Reason for Absence'] <= 21),1,0))
df_preprocessed.insert(0, 'Reason_2', np.where((df_preprocessed['Reason for Absence'] >= 15) & (df_preprocessed['Reason for Absence'] <= 17),1,0))
df_preprocessed.insert(0, 'Reason_1', np.where((df_preprocessed['Reason for Absence'] >= 1) & (df_preprocessed['Reason for Absence'] <= 14),1,0))
df_preprocessed = df_preprocessed.drop('Reason for Absence', axis=1)
df_preprocessed.head()
```

```python
# Extract month and day
df_preprocessed.insert(5, 'Day of Week', pd.to_datetime(df_preprocessed['Date'], format='%d/%m/%Y').dt.dayofweek)
df_preprocessed.insert(5, 'Month', pd.to_datetime(df_preprocessed['Date'], format='%d/%m/%Y').dt.month)
df_preprocessed = df_preprocessed.drop('Date', axis=1)
df_preprocessed.head()
```

```python
# Reformat education column
df_preprocessed['Education'] = np.where(df_preprocessed['Education']==1, 0, 1)
df_preprocessed.head()
```

```python
#Export csv
df_preprocessed.to_csv('Absenteeism_preprocessed.csv', index=False)
```
