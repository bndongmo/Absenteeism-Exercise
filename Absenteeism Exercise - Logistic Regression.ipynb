```python
import pandas as pd
import numpy as np
```

Load the Data

```python
data_preprocessed = pd.read_csv('Absenteeism_preprocessed.csv')
data_preprocessed.head()
```

Determine excessive amount of absenteeism

```python
median = data_preprocessed['Absenteeism Time in Hours'].median()
print(f'Median absent time is {median} hours')
targets = np.where(data_preprocessed['Absenteeism Time in Hours']>median,1,0)
targets
```

```python
data_preprocessed['Excessive Absenteeism'] = targets
data_preprocessed.head()
```

```python
targets.sum() / targets.shape[0]
```

```python
data_with_targets = data_preprocessed.drop('Absenteeism Time in Hours', axis=1)
data_with_targets.head()
```

Select inputs

```python
unscaled_inputs = data_with_targets.iloc[:,:-1]
unscaled_inputs
```

Scale the data

```python
from sklearn.preprocessing import StandardScaler
absenteeism_scaler = StandardScaler()
unscaled_inputs = unscaled_inputs.drop(['Daily Work Load Average', 'Day of Week', 'Distance to Work'], axis=1)
absenteeism_scaler.fit(unscaled_inputs)
scaled_inputs = absenteeism_scaler.transform(unscaled_inputs)
scaled_inputs
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(scaled_inputs, targets, train_size=.8, random_state=20)
```

Logistic regression with sklearn

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

```python
reg = LogisticRegression()
reg.fit(X_train, y_train)
```

```python
reg.score(X_train, y_train)
```

Manually check accuracy

```python
model_outputs = reg.predict(X_train)
```

```python
misclassifications = sum(abs(model_outputs-y_train))
samples = len(X_train)
print(misclassifications)
print((samples-misclassifications)/samples)
```

Intercept and coefficients

```python
feature_names = unscaled_inputs.columns.values
summary_table = pd.DataFrame(columns=['Feature Name'], data = feature_names)
summary_table['Coefficient'] = np.transpose(reg.coef_)
summary_table
```

```python
summary_table.index = summary_table.index+1
summary_table.loc[0] = ['Intercept', reg.intercept_.item()]
summary_table = summary_table.sort_index()
summary_table
```

Reason 4- Light diseases

```python
summary_table.sort_values('Coefficient', ascending=False)
```

Test model

```python
reg.score(X_test, y_test)
```

```python
predicted_proba = reg.predict_proba(X_test)
predicted_proba[:,1]
```

Save model

```python
import pickle
with open('model', 'wb') as file:
    pickle.dump(reg, file)
with open('scaler', 'wb') as file:
    pickle.dump(absenteeism_scaler, file)
```
